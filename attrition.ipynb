```python
from sklearn.model_selection import train_test_split
X_scale_train, X_scale_test, y_attr_train, y_attr_test, y_dept_train, y_dept_test = train_test_split(X_train_scaled, y_attrition, y_department)

```

```python
input_layer = layers.Input(shape=(X_train_scaled.shape[1],), name='input')

# Shared hidden layers
shared_layer1 = layers.Dense(64, activation='relu', name='shared1')(input_layer)
shared_layer2 = layers.Dense(128, activation='relu', name='shared2')(shared_layer1)
```

```python
deptartment_hidden = layers.Dense(32, activation='softmax', name='deptartment_hidden')(shared_layer2)

# Branch for color prediction
attrition_hidden = layers.Dense(32, activation='sigmoid', name='attrition_hidden')(shared_layer2)
dept_output = layers.Dense(3, activation='softmax', name='dept_output')(deptartment_hidden)

# Branch for color prediction
attrition_output = layers.Dense(2, activation='sigmoid', name='attrition_output')(attrition_hidden)

```

```python
dept_output
```

```python
model = Model(inputs=input_layer, outputs=[dept_output, attrition_output])

# Compile the model
model.compile(optimizer='adam',
              loss={'dept_output': 'categorical_crossentropy', 'attrition_output': 'categorical_crossentropy'},
              metrics={'dept_output': 'accuracy', 'attrition_output': 'accuracy'})

# Display the model summary
model.summary()

# Create the model


# Compile the model


# Summarize the model
```
